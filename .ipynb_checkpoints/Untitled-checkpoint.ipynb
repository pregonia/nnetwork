```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
tmax = 8
dt = 0.01
t = np.arange(0, tmax, dt)
tau = 1
tau2 = 3
ts = 1
tr = t[int(ts/dt):len(t)] - (ts-dt)
```

```python
gal = np.zeros(len(t))
galp = tau / np.exp(1)
```

```python
gal[int(ts/dt):len(t)] = tr * np.exp(-tr/tau) / galp
```

```python
plt.plot(gal)
plt.show()
```

```python
# Normalized concentration of the postsynaptic transmitter-receptor complex
# Presynaptic membrane potential function
def fvpre(V_pre):
    th_syn = 0
    return 1./(1+np.exp(-(V_pre - th_syn)/2))
```

```python
# Gating variable equation
from scipy.integrate import odeint

def V_pre(t):
    # Simulate a neuron spike
    V_pre_stim = np.zeros(len(t))
    stim = np.arange(-4,1,0.5)
    V_pre_stim[100:110] = 40*np.exp(stim-1)

    return V_pre_stim[t]

def dsdt(s, t, V_pre):
    alpha = 12
    beta = 0.1

    return alpha*fvpre(V_pre)*(1-s) - beta*s

# plt.figure()
# plt.plot(V_pre)

# s_0 = np.array([0.0] * len(t))

s_all = []
for V_pre, ts in zip(V_pre_stim, range(len(t))):
    s_0 = 0.0
    s_s = odeint(dsdt, s_0, ts, args=(V_pre,))
    print V_pre, ts, s_s
    s_all.append(s_s)
#     print s_s, s_all.shape, V_pre
#     print s_s.shape

s_all = np.squeeze(s_all)
s_all.shape

# plt.figure()
plt.plot(s_all)
plt.show()



# # plt.figure()
# # plt.plot(V_pre)

# s_0 = np.array([0.0] * len(t))

# # s_0 = 0.0
# V_pre = V_pre_all
# odeint(dsdt, s_0, t, args=(V_pre))

# # s_s = []
# # for V_pre in V_pre_all:
# #     s_s.append(odeint(dsdt, s_0, t, args=(V_pre)))

# # s_s = np.array(s_s)

# # plt.figure()
# # plt.plot(s_s[:,100])
# # plt.show()
```
