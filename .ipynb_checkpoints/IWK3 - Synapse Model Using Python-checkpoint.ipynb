# INTERNSHIP WEEK 3: Synapse Model

This notebook implements the synapse model by Wang and Buzsáki [1].

# Model

We followed Wang and Buzsáki's synapse model, which is defined by the following:

<center> 
$I_{syn} = g_{syn}s(V - E_{syn})$
<center>

where 

$I_{syn}$ is the synaptic current

$g_{syn}$ is the maximal synaptic conductance

$V$ is the membrane potential

$E_{syn}$ is the reverse potential
<br><br>

And the gating variable $s$ is defined by:

<center> 
$\frac{ds}{dt} = \alpha F(V_{pre})(1-s) - \beta s$
<center>

where 

$\alpha$ is the channel opening rate

$\beta$ is the channel closing rate

$V_{pre}$ is the presynaptic membrane potential, and

<center> 
$F(V_{pre}) = \dfrac{1}{1+e^{-(V_{pre} - \theta_{syn})/2}}$
<center>

# Parameter settings
In their paper, Wang and Buzsáki used the following values:

$g_{syn} = 0.1 \, mS/cm^2$

$E_{syn} = -75 \, mV$

$\theta_{syn} = 0 \, mV$

$\alpha = 12 \, msec^{-1}$

$\beta = 0.1 \, msec^{-1}$

# Implementation
The following code implements the synapse model in Python.

First, we generate a sample pre-synaptic voltage $V\_pre\_samp$ with a single pulse.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

import seaborn as sns

cp = sns.color_palette("hls", 7)
sns.set_palette(cp)
sns.set_color_codes()


# Constants
g_syn = 0.1  # mS/cm^2, maximal synaptic conductance
E_syn = -75  # mV, reversal potential
th_syn = 0   # mV, synaptic constant

al = 12      # 1/msec, channel opening rate
bt = 0.1     # 1/msec, channel closing reate

dt = 0.1     # msec, time step for RK4

# Sample input signal
t = np.arange(0, 1, 0.001)
N = len(t)

stim = np.arange(-4,1,0.5)
stim_2 = np.arange(0,20,0.1)
V_pre_samp = np.zeros(N)
V_pre_samp[:] = -64
V_pre_samp[100:110] = 163*np.exp(stim-1) - 64
V_pre_samp[110:310] = -0.5*np.exp(1-stim_2) - 64

plt.figure(figsize=(18, 4))
plt.plot(V_pre_samp, 'r', label='V_pre')
plt.ylabel('mV')
plt.xlabel('ms')
plt.legend()
plt.show()
```

Next, we implement the synapse model equations. <br>
We solve the gating variable ODE $ds/dt$ using 4th order Runge-Kutta method.

```python
# Numerical method
def rk4(f, t, X, V):
    ''' Solves the ODE using 4th order Runge-Kutta '''
    k1 = f(t, X, V)
    k2 = f(t+(dt/2), X+(dt*k1/2), V)
    k3 = f(t+(dt/2), X+(dt*k2/2), V)
    k4 = f(t+dt, X+(dt*k3), V)
    y = X + dt*(k1 + (2*k2) + (2*k3) + k4)/6

    return y


# Equations
def f_I_syn(s, V):
    ''' Returns the synaptic current '''
    return g_syn * s * (V-E_syn)

def f_V_pre(V_pre):
    ''' Returns the normalized concentration of
        postsynaptic transmitter-receptor complex '''
    return 1. / (1 + np.exp((th_syn-V_pre)/2))

def f_dsdt(t, s, V_pre):
    ''' Returns the gating variable '''
    exc = al * f_V_pre(V_pre) * (1-s)
    inh = bt * s
    dsdt =  exc - inh 
    return dsdt


# Run the simulation
si = 0 # initial value for s

s_samp, I_syn_samp = [], []
for ti, V_pre in zip(t, V_pre_samp):
    sj = rk4(f_dsdt, ti, si, V_pre)
    Ij = f_I_syn(sj, V_pre)

    s_samp.append(sj)
    I_syn_samp.append(Ij)

    si = sj
    
s_samp = np.array(s_samp)
I_syn_samp = np.array(I_syn_samp)


# Plot the results
plt.figure(figsize=(18, 10))
ax1 = plt.subplot(311)
plt.plot(V_pre_samp, 'r', label='V_pre')
plt.ylabel('$V_{pre}$ ($mV$)')

plt.subplot(312, sharex=ax1)
plt.plot(s_samp, 'g', label='S')
plt.ylabel('$s$')

plt.subplot(313, sharex=ax1)
plt.plot(-I_syn_samp, 'b', label='I_syn')
plt.ylabel('$I_{syn}$ ($mA$)')
plt.xlabel('t (ms)')
plt.show()
```

The plot shows the relation of the presynaptic voltage to the synaptic current. <br>

An excitation on the presynaptic neuron (pulse in $V_{pre}$) elicits an inhibitory postsynaptic current ($I_{syn}$).

<br> <br>
This is also the reason why we choose Wang and Buzsáki's synapse model. 

Unlike other models which rely on the presence of pulses in the presynaptic neuron, this model is dependent on the presynaptic voltage instead. Thus, this model is more robust since it doesn't have to monitor the incoming signal for pulses.
